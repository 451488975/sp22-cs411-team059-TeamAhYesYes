```python
df = pd.read_csv('train.csv')
df
```

```python
df_concat = df[['AnimalID', 'AnimalType', 'AgeuponOutcome', 'Breed', 'Color']]
df_concat
```

```python
uni = df['AgeuponOutcome'].unique()
print(len(uni))
uni
```

```python
age = df['AgeuponOutcome'].fillna('')
age = age.tolist()
```

```python
age[0]
```

```python
age_in_days = []
count = 0
for i in age:
    if i != '':
        temp = i.split(' ')
        temp_num = int(temp[0])
        temp_unit = temp[1]
        if temp_unit[:4] == 'year':
            day = temp_num * 365
        elif temp_unit[:4] == 'week':
            day = temp_num * 7
        elif temp_unit[:5] == 'month':
            day = temp_num * 30
        else:
            day = temp_num
    else:
        print(count)
        day = None
    count += 1
    age_in_days.append(day)
age_in_days
```

```python
len(age_in_days)
```

```python
df['AgeInDays'] = np.array(age_in_days)
```

```python
df_concat = df[['AnimalType', 'AgeInDays', 'Breed', 'Color']]
df_concat
```

```python
df_concat.to_csv('pet.csv')
```
